```python
#print a triangle pattern

i=1
while i<=5:
    j=1
    while j<=i:
        print("*", end="")
        j+=1
    print()
    i+=1
```

```python
#print number pattern

i=1
while i<=5:
    j=1
    while j<=i:
        print(i, end="")
        j+=1
    print()
    i+=1
```

```python
#Print String pattern
a="PYTHON"
i=1
while i<=len(a):
    print(a[:i])
    i+=1
```

```python
#Reverse the character
a= "PYTHON"
i= len(a)-1
while i>=0:
    print(a[i], end="")
    i-=1
```

```python
#Count the digits

a=12345
count=0
while a>0:
    a=a//10
    count+=1
print(count)
```

```python
#Find the largest number in a list

a=[98,76,55,112,72,90]
i=0
largest_number= a[i]

while i<len(a)-1:
    i+=1
    if a[i]>largest_number:
        largest_number=a[i]
print(largest_number)
    
```
